```python
import pandas as pd
```

```python
df = pd.read_csv('/home/prince/Data science/01_Practice/Pandas/00/hotel_bookings.csv')
```

```python
df.columns
```

```python
df.rename(columns = {'adults': 'num_adults', 'childrens':'num_childrens', 'babies' : 'num_babies'}, inplace=True)
```

```python
df['agent'].fillna(-1, inplace=True)
```

```python
df.isnull().sum() * 100/len(df)
```

```python
df[df['children'].isna()]
df.dropna(subset=['children'], inplace=True)
```

```python
df.drop(columns=['company'], inplace=True)
```

```python
df['country'].fillna('unknown', inplace=True)
```

```python
df.dtypes
```

```python
df = df.astype({'children':'int', 'is_repeated_guest':'boolean', 'is_canceled':'boolean', 'agent': 'int'})
df.dtypes
```

```python
labels = ['0-100', '101-200','201-300','301-400','401-500','501-600','601-700','701-800']
bins = [0,100, 200, 300, 400, 500, 600, 700, 800]
df['lead_time_bins']= pd.cut(df['lead_time'], labels=labels, bins=bins)
df[['lead_time','lead_time_bins']]
df.drop(columns='lead_time')
df.rename(columns= {'lead_time_bins': 'lead_time'}, inplace=True)
```

```python
df['arrival_date'] = (
    df['arrival_date_year'].astype(str) + '-' +
    df['arrival_date_month'].astype(str) + '-' +
    df['arrival_date_day_of_month'].astype(str)
)
```

```python
df = df.drop(['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month'], axis=1)
df[['arrival_date']]
```

```python
df.loc[df.duplicated(keep=False)]
df.drop_duplicates(keep='first', inplace=True)
```
